```python
import pandas as pd
import numpy as np
```

```python
p=pd.read_csv('product.csv')
c=pd.read_csv('custom.csv')
m=pd.read_csv('master.csv')
s1=pd.read_csv('search1.csv')
s2=pd.read_csv('search2.csv')
s=pd.read_csv('session.csv')
```

```python
p['PD_BUY_AM']=p['PD_BUY_AM'].apply(lambda x: x.replace(",",""))
p['PD_BUY_AM']=p['PD_BUY_AM'].astype(int)
p[p['PD_BUY_CT'] == '1,440'] = 1440
p[p['PD_BUY_CT'] == '1,387'] = 1387
p['PD_BUY_CT']=p['PD_BUY_CT'].astype(int)
```

```python
p["PD_BUY_TOT"] = 0
p["PD_BUY_TOT"] = p['PD_BUY_AM'] * p['PD_BUY_CT']
```

금액 분류 작업

```python
ps=p.groupby("CLNT_ID")['PD_BUY_TOT'].sum()
ps=pd.DataFrame(ps)
```

0원에서 10만원까지 검색건수

```python
ps[(ps.PD_BUY_TOT>0) & (ps.PD_BUY_TOT<=100000)].to_csv('ps010.txt')
ps010=pd.read_csv('ps010.txt')
cps010=pd.merge(c,ps010)
s1cps010=pd.merge(s1,cps010)
s1cps010.KWD_NM.value_counts().head(10)
```

10만원에서 50만원까지 검색건수

```python
ps[(ps.PD_BUY_TOT>100000) & (ps.PD_BUY_TOT<=500000)].to_csv('ps1050.txt')
ps1050=pd.read_csv('ps1050.txt')
cps1050=pd.merge(c,ps1050)
s1cps1050=pd.merge(s1,cps1050)
s1cps1050.KWD_NM.value_counts().head(10)
```

50만원 이상 검색건수

```python
ps[ps.PD_BUY_TOT>500000].to_csv('psov50.txt')
psov50=pd.read_csv('psov50.txt')
cpsov50=pd.merge(c,psov50)
s1cpsov50=pd.merge(s1,cpsov50)
s1cpsov50.KWD_NM.value_counts().head(10)
```

```python
ps[(ps.PD_BUY_TOT>0) & (ps.PD_BUY_TOT<=100000)].to_csv('ps010.txt')
ps010=pd.read_csv('ps010.txt')
pps010=pd.merge(p,ps010)
pps010['PD_BRA_NM']=pps010['PD_BRA_NM'].astype(str)
pps010['PD_BRA_NM']=pps010['PD_BRA_NM'].apply(lambda x: x.replace("[",""))
pps010['PD_BRA_NM']=pps010['PD_BRA_NM'].apply(lambda x: x.replace("]",""))
pps010.PD_BRA_NM.value_counts().head(10)
```

```python
ps[(ps.PD_BUY_TOT>100000) & (ps.PD_BUY_TOT<=500000)].to_csv('ps1050.txt')
ps1050=pd.read_csv('ps1050.txt')
pps1050=pd.merge(p,ps1050)
pps1050['PD_BRA_NM']=pps1050['PD_BRA_NM'].astype(str)
pps1050['PD_BRA_NM']=pps1050['PD_BRA_NM'].apply(lambda x: x.replace("[",""))
pps1050['PD_BRA_NM']=pps1050['PD_BRA_NM'].apply(lambda x: x.replace("]",""))
pps1050.PD_BRA_NM.value_counts().head(10)
```
